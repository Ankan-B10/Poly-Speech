```python
import boto3
#aws_access_key_id='AKIAZM3DNNHINV6YW355',aws_secret_access_key='k3MB8mhWDA9DCQ3DJAK5bnrmegI9YJjmUpL56Gen'
polly = boto3.client('polly',region_name='us-east-1',aws_access_key_id='Enter your access id',aws_secret_access_key='Enter your access secret key')
```

```python
ssml_text = f"""
<speak>
    <prosody rate="slow">
    Hello, I am Ankan Bera, 3rd yr student of Haldia Institute of Technology.
    </prosody>
</speak>
"""
```

```python
response = polly.synthesize_speech(Text=ssml_text, OutputFormat="mp3",VoiceId="Aditi",TextType="ssml")
if "AudioStream" in response:
    with response["AudioStream"] as stream:
        output_file = "speech_with_speed.mp3"
        try:
            with open(output_file, "wb") as file:
                file.write(stream.read())
            print(f"Audio saved as {output_file}")
        except IOError as error:
            print(error)
else:
    print("Could not stream audio")
```
